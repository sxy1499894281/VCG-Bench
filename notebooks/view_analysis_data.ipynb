## 四、数据可视化预览


```python
# 4.1 可视化Task1模型对比（如果数据存在）
if 'task1_by_model' in locals() and task1_by_model is not None:
    # 选择主要指标进行可视化
    metrics = ['style_consistency_score_mean', 'codevqa_mean', 'execution_success_rate_mean', 'siglip_score_mean']
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    axes = axes.flatten()
    
    for idx, metric in enumerate(metrics):
        if metric in task1_by_model.columns:
            data = task1_by_model[['model', metric]].sort_values(metric, ascending=False)
            axes[idx].barh(data['model'], data[metric])
            axes[idx].set_title(metric.replace('_', ' ').title())
            axes[idx].set_xlabel('Score')
            axes[idx].tick_params(axis='y', labelsize=8)
    
    plt.tight_layout()
    plt.show()
else:
    print("Task1模型数据不存在，跳过可视化")
```

```python
# 4.2 可视化CodeVQA按问题类型表现（如果数据存在）
if 'task1_codevqa' in locals() and task1_codevqa is not None:
    # 选择几个主要模型进行对比
    top_models = task1_codevqa['model'].unique()[:5]  # 选择前5个模型
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    x = np.arange(len(['counting', 'identification', 'relationship']))
    width = 0.15
    
    for i, model in enumerate(top_models):
        model_data = task1_codevqa[task1_codevqa['model'] == model]
        accuracies = []
        for qtype in ['counting', 'identification', 'relationship']:
            qtype_data = model_data[model_data['question_type'] == qtype]
            acc = qtype_data['accuracy'].values[0] if len(qtype_data) > 0 else 0
            accuracies.append(acc)
        ax.bar(x + i*width, accuracies, width, label=model.split('/')[-1])
    
    ax.set_xlabel('Question Type')
    ax.set_ylabel('Accuracy')
    ax.set_title('CodeVQA Performance by Question Type')
    ax.set_xticks(x + width * 2)
    ax.set_xticklabels(['counting', 'identification', 'relationship'])
    ax.legend()
    ax.set_ylim([0, 1])
    
    plt.tight_layout()
    plt.show()
else:
    print("Task1 CodeVQA数据不存在，跳过可视化")
```
